DECODE METADATA DE LOS ARCHIVOS .fff


```python
import matplotlib.pyplot as plt
from flirpy.io.fff import Fff

# Lista de rutas de archivos FFF
file_paths = [
    '/content/CTP_RUMHIR_20231208_0900(1)(1).fff',
    '/content/CTP_RUMHIR_20231208_0905(1).fff',
    '/content/CTP_RUMHIR_20231208_0910.fff',
    '/content/CTP_RUMHIR_20231208_0915.fff',
    '/content/CTP_RUMHIR_20231208_0920.fff',
    '/content/CTP_RUMHIR_20231208_0925(1).fff'
]

for i, file_path in enumerate(file_paths):
    # Cargar un archivo FFF
    fff_reader = Fff(file_path)

    # Leer la imagen térmica
    thermal_image = fff_reader.get_image()

    # Crear una nueva figura para cada imagen térmica
    plt.figure(figsize=(6, 5))
    plt.imshow(thermal_image, cmap='inferno')
    plt.colorbar(label='Temperatura (°C)')
    plt.title(f'Imagen Térmica {i+1}')
    plt.xlabel('Pixel X')
    plt.ylabel('Pixel Y')
    plt.show()

```
