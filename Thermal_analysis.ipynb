DECODE METADATA DE LOS ARCHIVOS .fff


```python
import matplotlib.pyplot as plt
from flirpy.io.fff import Fff

# Ruta del archivo FFF
file_path = '/content/foto.fff'  # Cambia esta ruta por la de tu archivo

# Cargar un archivo FFF
fff_reader = Fff(file_path)

# Leer la imagen térmica
thermal_image = fff_reader.get_image()

# Mostrar la imagen térmica
plt.imshow(thermal_image, cmap='inferno')
plt.colorbar(label='Temperatura (°C)')
plt.title('Imagen Térmica')
plt.xlabel('Pixel X')
plt.ylabel('Pixel Y')
plt.show()
```
